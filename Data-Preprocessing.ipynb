```python
import re

text = "Invoice Number: INV123456, Total Amount: $500.00, Date: 2023-01-15"

# Extracting the invoice number using a regular expression
invoice_number_match = re.search(r'Invoice Number: (\w+).*Date: (\d{4}-\d{2}-\d{2})', text)

if invoice_number_match:
    invoice_number = invoice_number_match.group(1)
    date= invoice_number_match.group(2)
    print("Invoice Number:", invoice_number)
    print("date:", date)
else:
    print("Invoice number not found.")
```
